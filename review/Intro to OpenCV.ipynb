## Opening some images

```python
img = cv2.imread("cats_and_dogs_filtered/train/cats/cat.1.jpg")
```

```python
plt.imshow(img)
```

Do you think the colors seem a bit off? Let's fix that with the following code.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
print(img.shape)
```

```python
print(type(img))
```

## Resizing an image

```python
img = cv2.resize(img, (128, 128))
```

```python
plt.imshow(img)
```

```python
img.shape
```

## Cropping an Image

 OpenCV doesn't have an internal function for cropping. However, since images read in using OpenCV are Numpy arrays, cropping an image by ourselves is easy.

```python
cropped_img = img[20:80, 20:80]
```

```python
plt.imshow(cropped_img)
```

```python
cropped_img.shape
```

#### These functions are the minimum you need to know to work with images in machine learning. However, we've included some more functions shown below.

## Flipping an image

```python
img1 = cv2.flip(img, 0)
```

```python
plt.imshow(img1)
```

```python
img1 = cv2.flip(img, 1)
```

```python
plt.imshow(img1)
```

## Rotating an image

```python
img1 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
```

```python
plt.imshow(img1)
```

```python
img1 = cv2.rotate(img, cv2.ROTATE_180)
```

```python
plt.imshow(img1)
```

## Drawing boxes on images

cv2.rectangle(< img variable >, (x1, y1), (x2, y2), < color >, < size of lines >)

```python
img = cv2.imread("cats_and_dogs_filtered/train/cats/cat.1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (128, 128))

img1 = cv2.rectangle(img, (20, 20), (85, 100), (255, 0, 0), 1)
```

```python
plt.imshow(img1)
```
