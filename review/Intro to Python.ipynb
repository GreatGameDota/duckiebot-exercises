# If Else Statements

if (*expression*):  
*tab* *code*  
else:  
*tab* *code*

````python
if (False):
  print("True")
else:
  print("False")
````

````python
if (booleon_variable1):
  print("Variable 1 is True!")
else:
  print("Variable 1 is False!")

if (booleon_variable2):
  print("Variable 2 is True!")
else:
  print("Variable 2 is False!")
````

# Elif Statements

Functions the same as other languages' ```else if``` statements.

if (*expression*):  
*tab* *code*  
elif (*expression*):  
*tab* *code*  
else:  
*tab* *code*

````python
if (False):
  print("False")
elif (True):
  print("True")
else:
  print("False")
````

````python
booleon_variable1 = True
booleon_variable2 = False

if (booleon_variable1):
  print("Variable 1 is True!")
elif (booleon_variable2):
  print("Variable 2 is True!")
else:
  print("Variable 1 and 2 are False!")
````

````python
booleon_variable1 = False
booleon_variable2 = True

if (booleon_variable1):
  print("Variable 1 is True!")
elif (booleon_variable2):
  print("Variable 2 is True!")
else:
  print("Variable 1 and 2 are False!")
````

````python
booleon_variable1 = False
booleon_variable2 = False

if (booleon_variable1):
  print("Variable 1 is True!")
elif (booleon_variable2):
  print("Variable 2 is True!")
else:
  print("Variable 1 and 2 are False!")
````

# Functions

def <*function name*>(*parameters*):  
*tab* *code*

````python
def example_function1():
  print("Hello!")
````

````python
example_function1()
````

#### Input Parameters

````python
def example_function2(parameter1):
  print(parameter1)
````

````python
example_function2("Hello!")
````

````python
example_function2(100)
````

````python
def example_function3(param1, param2):
  print(f'Param1: {param1} Param2: {param2}')
````

````python
example_function3("Hello!", 100)
````

#### Return Values

````python
def example_function4(param1):
  return param1 * 2
````

````python
print(example_function4(100))
````

````python
def example_function5(param1, param2):
  return param1 + param2
````

````python
print(example_function5(10, 20))
````

````python
def example_function6(param1, param2):
  return param1 + 1, param2 + 1
````

````python
print(example_function6(10, 20))
````

````python
print(example_function6(10, 20)[0])
````

````python
print(example_function6(10, 20)[1])
````

````python
def example_function7(param1, param2, param3):
  return param1 + 1, param2 + 1, param3 + 1
````

````python
print(example_function7(10, 20, 30))
````

````python
type(example_function7(10, 20, 30))
````
