# NumPy is an important Python package for multi-dimensional array storage and manipulation.

````python
import numpy as np
````

NumPy is extremely handy for working with machine learning, specifically neural networks. Using it gives you a quick way of creating multi-dimensional arrays and performing operations on them.

## Creating a NumPy array

np.zeros(< *shape* >)

This function creates a NumPy array filled entirely with zeros of the shape, the number of rows and columns, inputted into the function. (You can also use ```np.ones()``` or ```np.empty()```)

````python
arr = np.zeros((100, 100))
````

````python
# You can directly print the array with Python
print(arr)
````

You can also create an array manually:

````python
arr = np.array([0, 1, 2, 3])
````

````python
arr
````

````python
arr1 = [0, 1, 2]

arr1 = np.array(arr1)
````

This function can also be used for casting a normal Python array into a NumPy array

## Random initialization of an array

np.random.randint(< *lower bound* >, < *upper bound* >, < *shape* >)

````python
random_arr = np.random.randint(0, 10, (10, 10))
````

````python
random_arr
````

## Ranged array

Like ```range()``` in Python you can create a ranged numpy array with ```np.arange```.

````python
ranged_arr = np.arange(0, 10)
````

````python
ranged_arr
````

## Shape of an array

This also functions as a way to get the size of an array

````python
arr.shape
````

````python
arr.shape[0]
````

## Indexing an array

Indexing a NumPy array works the same as a normal Python array

````python
arr = np.zeros((100, 100))

print(arr[0])
````

````python
print(arr[0][0])
````

# Basic high level operations and arithmetic

This notebook won't cover every operation, just the basic ones in the NumPy library. You are free to search up documentation to learn about all the functionality of this library.

## Add two arrays

````python
arr1 = np.array([0, 1, 2, 3])
arr2 = np.array([1, 1, 1, 1])

print(arr1)
print(arr2)
````

````python
arr3 = arr1 + arr2
````

````python
print(arr3)
````

## Subtract two arrays

````python
arr4 = arr1 - arr2
````

````python
print(arr4)
````

## Multiply two arrays

````python
arr5 = arr1 * arr2
````

````python
print(arr5)
````

## Dot product of two arrays/matrices

````python
arr6 = np.dot(arr1, arr2)
````

````python
print(arr6)
````

## Cross product of two arrays/matrices

````python
arr1 = np.array([2, 3, 4])
arr2 = np.array([5, 6, 7])

arr7 = np.cross(arr1, arr2)
````

````python
print(arr7)
````
